```python
import pandas as pd
import numpy as np
```

```python
dfx = pd.read_csv('Diabetes_XTrain.csv')
dfy = pd.read_csv('Diabetes_YTrain.csv')
df = pd.read_csv('Diabetes_Xtest.csv')
df.head()
```

```python
dfx.head()
```

```python
dfy.head()
```

```python
X = dfx.values
Y = dfy.values
print(X.shape)
print(Y.shape)
```

```python
def dist(x1,x2):
    return np.sqrt(sum((x1-x2)**2))
```

```python
def knn(X,Y,querypoint,k=5):
    vals = []
    m = X.shape[0]
    for i in range(m):
        d = dist(X[i],querypoint)
        vals.append((d,Y[i]))
    vals = sorted(vals)
    vals = vals[:k]
    vals = np.array(vals)
    vals2=vals[:, 1]
    s = sum(vals2)
    
    if s > 3:
        return 1
    else:
        return 0
```

```python
test = df.values
test.shape
```

```python
k = test.shape[0]
#print(k)
```

```python
final_label = np.full((192,1), 0)

for i in range(k):
    y_pred = knn(X, Y, test[i], 5)
    final_label[i][0] = y_pred
print(final_label)
```

```python
sub = pd.DataFrame(final_label, columns = ["Outcome"])
sub.head()
```

```python
sub = sub.to_csv("prediction.csv", index = False)
```
